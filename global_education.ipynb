```python
import pandas as pd

# Carregar o arquivo CSV
file_path = 'Global_Education.csv'
# Tentar carregar o arquivo CSV usando uma codificação diferente
data = pd.read_csv(file_path, encoding='latin1')

# Exibir as primeiras linhas do dataframe para entender sua estrutura
data.head(), data.columns

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo do seaborn
sns.set(style="whitegrid")

# Gráfico de dispersão: Taxa de Desemprego vs. Taxa Bruta de Matrícula no Ensino Superior
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Gross_Tertiary_Education_Enrollment', y='Unemployment_Rate', data=data)
plt.title('Taxa de Desemprego vs. Taxa Bruta de Matrícula no Ensino Superior')
plt.xlabel('Taxa Bruta de Matrícula no Ensino Superior (%)')
plt.ylabel('Taxa de Desemprego (%)')
plt.savefig('desemprego_vs_terciario.png')
plt.show()

# Gráfico de dispersão: Taxa de Desemprego vs. Taxa de Alfabetização de Jovens (15-24 anos)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Youth_15_24_Literacy_Rate_Female', y='Unemployment_Rate', data=data, label='Feminino')
sns.scatterplot(x='Youth_15_24_Literacy_Rate_Male', y='Unemployment_Rate', data=data, label='Masculino')
plt.title('Taxa de Desemprego vs. Taxa de Alfabetização de Jovens (15-24 anos)')
plt.xlabel('Taxa de Alfabetização de Jovens (%)')
plt.ylabel('Taxa de Desemprego (%)')
plt.legend()
plt.savefig('desemprego_vs_alfabetizacao_jovens.png')
plt.show()

# Gráfico de barras: Taxa de Conclusão do Ensino Primário, Secundário Inferior e Superior por Gênero
conclusao_genero = data[['Completion_Rate_Primary_Male', 'Completion_Rate_Primary_Female',
                         'Completion_Rate_Lower_Secondary_Male', 'Completion_Rate_Lower_Secondary_Female',
                         'Completion_Rate_Upper_Secondary_Male', 'Completion_Rate_Upper_Secondary_Female']].mean()

conclusao_genero.index = ['Primário (Masculino)', 'Primário (Feminino)',
                          'Secundário Inferior (Masculino)', 'Secundário Inferior (Feminino)',
                          'Secundário Superior (Masculino)', 'Secundário Superior (Feminino)']

plt.figure(figsize=(14, 8))
conclusao_genero.plot(kind='bar', color=['blue', 'pink', 'blue', 'pink', 'blue', 'pink'])
plt.title('Taxa de Conclusão por Nível de Ensino e Gênero')
plt.xlabel('Nível de Ensino e Gênero')
plt.ylabel('Taxa de Conclusão (%)')
plt.xticks(rotation=20)
plt.savefig('conclusao_genero.png')
plt.show()

# Gráfico de linhas: Taxa de Desemprego em função da Taxa de Matrícula no Ensino Primário
plt.figure(figsize=(10, 6))
sns.lineplot(x='Gross_Primary_Education_Enrollment', y='Unemployment_Rate', data=data)
plt.title('Taxa de Desemprego em função da Taxa de Matrícula no Ensino Primário')
plt.xlabel('Taxa Bruta de Matrícula no Ensino Primário (%)')
plt.ylabel('Taxa de Desemprego (%)')
plt.savefig('desemprego_vs_primario.png')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Carregar shapefile de países para o mapa
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Renomear coluna para combinar com os dados
world = world.rename(columns={'name': 'Countries and areas'})

# Padronizar os nomes dos países
world['Countries and areas'] = world['Countries and areas'].str.strip()
data['Countries and areas'] = data['Countries and areas'].str.strip()

# Unir dados geográficos com dados de educação e desemprego
geo_data = world.set_index('Countries and areas').join(data.set_index('Countries and areas'), how='inner')

# Coropleta: Taxa de Desemprego por País
plt.figure(figsize=(15, 10))
geo_data.plot(column='Unemployment_Rate', cmap='OrRd', legend=True, missing_kwds={"color": "lightgrey"})
plt.title('Taxa de Desemprego por País')
plt.savefig('/mnt/data/coropleta_desemprego.png')
plt.show()
```
